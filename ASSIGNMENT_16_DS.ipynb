

### **1. Create a list called years_list, starting with the year of your birth, and each year thereafter until the year of your fifth birthday. For example, if you were born in 1980. the list would be years_list = [1980, 1981, 1982, 1983, 1984, 1985]..**  



**ANSWER**

**Example:** If you were born in **1980**, the list would be:  
```python
years_list = [1980, 1981, 1982, 1983, 1984, 1985]
```  
**Explanation:**  
- The list starts with the birth year (e.g., `1980`).  
- Each subsequent year represents one additional year of age.  
- The fifth birthday occurs in the **sixth** year of the list (since age `0` is the first year).  

**How to create it:**  
```python
birth_year = 1990  # Replace with your birth year
years_list = [birth_year + i for i in range(6)]  # [1990, 1991, ..., 1995]
```  

---

### **2. In which year in years_list was your third birthday? Remember, you were 0 years of age for your first year.?**
  
**Answer:**  
```python
third_birthday_year = years_list[3]  # Result: 1993 (if born in 1990)
```  
**Explanation:**  
- At index `0`: You are **0 years old** (birth year).  
- At index `3`: You turn **3 years old**.  

---

### **3. In the `years_list`, which year were you the oldest?**  
**Answer:**  
```python
oldest_year = years_list[-1]  # Last element (e.g., 1995)
```  
**Explanation:**  
- The last element represents the year of your **fifth birthday** (oldest age in the list).  

---

### **4. Make a list called `things` with these three strings: "mozzarella", "cinderella", "salmonella".**  
**Answer:**  
```python
things = ["mozzarella", "cinderella", "salmonella"]
```  
**Note:**  
- "mozzarella" is a cheese.  
- "cinderella" is a person (fairy tale character).  
- "salmonella" is a bacterium.  

---

### **5. Capitalize the element in `things` that refers to a person and then print the list. Did it change the element in the list?**  
**Answer:**  
```python
things[1] = things[1].capitalize()  # "cinderella" → "Cinderella"
print(things)  # Output: ['mozzarella', 'Cinderella', 'salmonella']
```  
**Explanation:**  
- `capitalize()` modifies the string in place.  
- The list now reflects the updated value ("Cinderella").  

---

### **6. Make a `surprise` list with the elements "Groucho," "Chico," and "Harpo."**  
**Answer:**  
```python
surprise = ["Groucho", "Chico", "Harpo"]
```  
**Note:**  
These are the names of the **Marx Brothers**, a famous comedy trio.  

---

### **7. Lowercase the last element of the `surprise` list, reverse it, and then capitalize it.**  
**Answer:**  
```python
last_element = surprise[-1].lower()      # "Harpo" → "harpo"
reversed_element = last_element[::-1]   # "harpo" → "oprah"
capitalized_element = reversed_element.capitalize()  # "oprah" → "Oprah"
surprise[-1] = capitalized_element      # Updates list to ["Groucho", "Chico", "Oprah"]
```  
**Result:**  
```python
print(surprise)  # Output: ['Groucho', 'Chico', 'Oprah']
```  
**Explanation:**  
- The last name "Harpo" is transformed into "Oprah" (as a playful reference to Oprah Winfrey).  

---

### **8. Make an English-to-French dictionary called `e2f` and print it.**  
**Given:**  
- dog → chien  
- cat → chat  
- walrus → morse  

**Answer:**  
```python
e2f = {"dog": "chien", "cat": "chat", "walrus": "morse"}
print(e2f)  # Output: {'dog': 'chien', 'cat': 'chat', 'walrus': 'morse'}
```  



### **9. Write the French word for walrus in your three-word dictionary `e2f`.**  
**Answer:**  
```python
french_walrus = e2f["walrus"]  # Result: "morse"
```  
**Explanation:**  
- The dictionary `e2f` maps English words to French.  
- Access the value for the key `"walrus"` to get the French translation.  

---

### **10. Make a French-to-English dictionary called `f2e` from `e2f`. Use the `items` method.**  
**Answer:**  
```python
f2e = {french: english for english, french in e2f.items()}
print(f2e)  # Output: {'chien': 'dog', 'chat': 'cat', 'morse': 'walrus'}
```  
**Explanation:**  
- `e2f.items()` returns pairs of English-French words.  
- Dictionary comprehension swaps keys (French) and values (English) to create `f2e`.  

---

### **11. Print the English version of the French word "chien" using `f2e`.**  
**Answer:**  
```python
english_word = f2e["chien"]  # Result: "dog"
print(english_word)  # Output: dog
```  

---

### **12. Make and print a set of English words from the keys in `e2f`.**  
**Answer:**  
```python
english_words = set(e2f.keys())  # Result: {'dog', 'cat', 'walrus'}
print(english_words)
```  
**Explanation:**  
- `e2f.keys()` gets the English words.  
- `set()` converts them to an unordered, unique collection.  

---

### **13. Make a multilevel dictionary called life. Use these strings for the topmost keys: &#39;animals&#39;, &#39;plants&#39;,and &#39;other&#39;. Make the &#39;animals&#39; key refer to another dictionary with the keys &#39;cats&#39;, &#39;octopi&#39;, and &#39;emus&#39;. Make the &#39;cats&#39; key refer to a list of strings with the values &#39;Henri&#39;, &#39;Grumpy&#39;, and &#39;Lucy&#39;.**
Make all the other keys refer to empty dictionaries.


**Answer:**  
```python
life = {
    'animals': {
        'cats': ['Henri', 'Grumpy', 'Lucy'],
        'octopi': {},
        'emus': {}
    },
    'plants': {},
    'other': {}
}
```  
**Structure:**  
- Top-level keys: `'animals'`, `'plants'`, `'other'`.  
- `'animals'` contains:  
  - `'cats'`: A list of cat names.  
  - `'octopi'` and `'emus'`: Empty dictionaries.  
- `'plants'` and `'other'` are empty.  

---

### **14. Print the top-level keys of `life`.**  
**Answer:**  
```python
print(life.keys())  # Output: dict_keys(['animals', 'plants', 'other'])
```  
**Alternative (as a list):**  
```python
print(list(life.keys()))  # Output: ['animals', 'plants', 'other']
```  

---

### **15. Print the keys for `life['animals']`.**  
**Answer:**  
```python
print(life['animals'].keys())  # Output: dict_keys(['cats', 'octopi', 'emus'])
```  
**Alternative (as a list):**  
```python
print(list(life['animals'].keys()))  # Output: ['cats', 'octopi', 'emus']
```  

---

### **16. Print the values for `life['animals']['cats']`.**  
**Answer:**  
```python
print(life['animals']['cats'])  # Output: ['Henri', 'Grumpy', 'Lucy']
```  

---
